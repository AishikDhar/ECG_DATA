```python
import os
import wfdb
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.fftpack import fft
from scipy.signal import find_peaks, butter, filtfilt

# Function to load raw ECG data from WFDB files
def load_raw_data(df, sampling_rate=100):
    data = [wfdb.rdsamp(f) for f in df.filename_lr]
    data = np.array([signal for signal, meta in data])
    return data

# Function to aggregate diagnostic superclass from SCP codes
def aggregate_diagnostic(y_dic, agg_df):
    temp = []
    for key in y_dic.keys():
        if key in agg_df.index:
            temp.append(agg_df.loc[key].diagnostic_class)
    return list(set(temp))

# Function to remove baseline drift using a high-pass filter
def remove_baseline(ecg_values, cutoff=0.5, fs=500, order=2):
    nyquist = 0.5 * fs
    normal_cutoff = cutoff / nyquist
    b, a = butter(order, normal_cutoff, btype='high', analog=False)
    filtered_ecg = filtfilt(b, a, ecg_values)
    return filtered_ecg

# Function to standardize ECG data (mean 0, std 1)
def standardize_data(ecg_values):
    mean = np.mean(ecg_values)
    std = np.std(ecg_values)
    standardized_ecg = (ecg_values - mean) / std
    return standardized_ecg

# Function to read, plot, and save ECG records
def read_plot_and_save_records(base_dir, num_records, output_dir):
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
        
    all_record_dataframes = []
        
    for i in range(num_records):
        record_num = str(i).zfill(5)  # Ensure the record number is zero-padded to 5 digits
        record_path = os.path.join(base_dir, record_num + '_lr')
        
        # Check for the .hea file existence (since .dat might not be present)
        if os.path.exists(record_path + '.hea'):
            record = wfdb.rdrecord(record_path)
            print(f"Plotting record {record_num}")
            wfdb.plot_wfdb(record=record, title=f'Record {record_num} from PTB-XL')

            # Convert the signal data to a pandas DataFrame
            df = pd.DataFrame(record.p_signal, columns=record.sig_name)
            print(df.head())

            # Write the DataFrame to a CSV file
            csv_filename = os.path.join(output_dir, f'R{record_num}_lr.csv')
            df.to_csv(csv_filename, index=False)
            
            # Append DataFrame to list for later use
            all_record_dataframes.append(df)
            
        else:
            print(f"Record {record_num} not found.")
    
    return all_record_dataframes

# Function to analyze the first ECG signal in a DataFrame
def analyze_first_ecg_signal(df):
    first_signal_name = df.columns[0]  # assuming the first column is the signal of interest
    print(f"Signal Name: {first_signal_name}")
    ecg_values = df[first_signal_name].values
    
    filtered_ecg = remove_baseline(ecg_values)
    standardized_ecg = standardize_data(filtered_ecg)
    
    plt.figure(figsize=(12, 6))
    plt.hist(standardized_ecg, bins=50, color='blue', alpha=0.7)
    plt.title('Distribution of Standardized ECG Values')
    plt.xlabel('ECG Value')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()

    std_dev = np.std(standardized_ecg)
    print(f'Standard Deviation of ECG Values: {std_dev}')

    mean_value = np.mean(standardized_ecg)
    median_value = np.median(standardized_ecg)
    print(f'Mean of ECG Values: {mean_value}')
    print(f'Median of ECG Values: {median_value}')
    
    return standardized_ecg

# Function for frequency domain analysis using FFT
def frequency_domain_analysis(standardized_ecg, fs=100):
    N = len(standardized_ecg)
    T = 1.0 / fs
    yf = fft(standardized_ecg)
    xf = np.linspace(0.0, 1.0/(2.0*T), N//2)

    plt.figure(figsize=(14, 7))
    plt.plot(xf, 2.0/N * np.abs(yf[:N//2]))
    plt.title('Frequency Domain Representation (FFT)')
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Amplitude')
    plt.grid(True)
    plt.show()

    return xf, yf, N

# Function for identifying peaks in the frequency domain and extracting required values
def peak_analysis_frequency_domain(xf, yf, N):
    peaks, _ = find_peaks(2.0/N * np.abs(yf[:N//2]), height=0)
    peak_freqs = xf[peaks]
    peak_amplitudes = 2.0/N * np.abs(yf[:N//2])[peaks]

    if len(peak_freqs) > 0:
        # Sort peaks by amplitude in descending order
        sorted_peak_indices = np.argsort(peak_amplitudes)[::-1]
        sorted_peak_freqs = peak_freqs[sorted_peak_indices]
        sorted_peak_amplitudes = peak_amplitudes[sorted_peak_indices]

        # Extract frequencies of maximum peak and second highest peak
        max_peak_freq = sorted_peak_freqs[0]
        second_peak_freq = sorted_peak_freqs[1] if len(sorted_peak_freqs) > 1 else None

        # Calculate ratio between f and 2f
        if second_peak_freq is not None and max_peak_freq != 0:
            ratio_f_2f = max_peak_freq / second_peak_freq
        else:
            ratio_f_2f = None

        return max_peak_freq, second_peak_freq, ratio_f_2f
    else:
        return None, None, None

# Function for noise analysis in the frequency domain
def noise_analysis(xf, yf, N, peak_amplitudes):
    noise_threshold = 0.1 * max(peak_amplitudes)
    noise = np.abs(yf) < noise_threshold

    plt.figure(figsize=(14, 7))
    plt.plot(xf, 2.0/N * np.abs(yf[:N//2]), label='Signal')
    plt.plot(xf[noise[:N//2]], 2.0/N * np.abs(yf[:N//2])[noise[:N//2]], 'r.', label='Noise')
    plt.title('Noise Analysis in ECG Signal')
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Amplitude')
    plt.legend()
    plt.grid(True)
    plt.show()

# Function for time domain analysis of RR intervals
def time_domain_peak_analysis(standardized_ecg, fs=500):
    threshold = 0.5 * max(standardized_ecg)
    peaks, _ = find_peaks(standardized_ecg, height=threshold)
    peak_distances = np.diff(peaks)
    std_dev_distances = np.std(peak_distances)
    
    mean_rr_intervals = np.mean(peak_distances / fs)
    median_rr_intervals = np.median(peak_distances / fs)
    std_dev_rr_intervals = np.std(peak_distances / fs)
    
    plt.figure(figsize=(14, 7))
    plt.plot(standardized_ecg, label='ECG Signal')
    plt.plot(peaks, standardized_ecg[peaks], "x", label='Peaks')
    plt.title('Peaks in ECG Signal')
    plt.xlabel('Sample Index')
    plt.ylabel('Amplitude')
    plt.legend()
    plt.grid(True)
    plt.show()

    print(f'Standard Deviation of Peak Distances: {std_dev_distances}')
    print(f'Mean RR Intervals: {mean_rr_intervals} seconds')
    print(f'Median RR Intervals: {median_rr_intervals} seconds')
    print(f'Standard Deviation RR Intervals: {std_dev_rr_intervals} seconds')

    return mean_rr_intervals, median_rr_intervals, std_dev_rr_intervals

# Load metadata from PTB-XL database
base_dir = 'C:/records100/00000/'
output_dir = 'C:/records100/CSV_Files/'

# Number of records to read, plot, and save
num_records = 5

# Read, plot, and save the records
record_dataframes = read_plot_and_save_records(base_dir, num_records, output_dir)

# List to store standardized Lead I signal values for each record
lead_I_values_list = []

# List to store RR interval statistics for each record
rr_intervals_list = []

# List to store Lead I signal details for each record
lead_I_details_list = []

# Analyze each record in record_dataframes
for i, df_record in enumerate(record_dataframes, start=1):
    print(f"\nAnalysis for Record {i}:")
    
    # Analyze first ECG signal in the DataFrame
    standardized_ecg = analyze_first_ecg_signal(df_record)
    
    # Append standardized Lead I signal values to the list
    lead_I_values_list.append(standardized_ecg)
    
    # Extract Lead I signal details
    lead_I_mean = np.mean(standardized_ecg)
    lead_I_median = np.median(standardized_ecg)
    lead_I_std_dev = np.std(standardized_ecg)
    
    # Store Lead I details in a dictionary
    lead_I_details_list.append({
        'Record': f'Record {i}',
        'Lead I Mean': lead_I_mean,
        'Lead I Median': lead_I_median,
        'Lead I Std Dev': lead_I_std_dev
    })

    # Convert the list of dictionaries to a DataFrame
    lead_I_details_df = pd.DataFrame(lead_I_details_list)
    # Frequency domain analysis
    xf, yf, N = frequency_domain_analysis(standardized_ecg)
    
    # Peak analysis in the frequency domain
    max_peak_freq, second_peak_freq, ratio_f_2f = peak_analysis_frequency_domain(xf, yf, N)
    
    # Time domain analysis of RR intervals
    mean_rr_intervals, median_rr_intervals, std_dev_rr_intervals = time_domain_peak_analysis(standardized_ecg)
    
    # Store results in DataFrame
    rr_intervals_list.append({
        'Record': f'Record {i}',
        'Mean RR Interval': mean_rr_intervals,
        'Median RR Interval': median_rr_intervals,
        'Std Dev RR Interval': std_dev_rr_intervals,
        'Max Peak Frequency': max_peak_freq,
        'Second Peak Frequency': second_peak_freq,
        'Ratio f / 2f': ratio_f_2f
    })
# Convert the list of arrays to a DataFrame
lead_I_values_df = pd.DataFrame(lead_I_values_list)

# Convert the list of dictionaries to a DataFrame
rr_intervals_df = pd.DataFrame(rr_intervals_list)

# Print the DataFrame with RR interval statistics and frequency domain analysis results
print("\nRR Interval and Frequency Domain Analysis:")
print(rr_intervals_df[['Record', 'Mean RR Interval', 'Median RR Interval', 'Std Dev RR Interval', 
                       'Max Peak Frequency', 'Second Peak Frequency', 'Ratio f / 2f']])

# Print the DataFrame with standardized Lead I signal values
print("\nStandardized Lead I Signal Values:")
print(lead_I_values_df)

print("\nLead I Signal Details:")
print(lead_I_details_df[['Record', 'Lead I Mean', 'Lead I Median', 'Lead I Std Dev']])
```
